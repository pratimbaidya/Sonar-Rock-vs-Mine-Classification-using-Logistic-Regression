# Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection and Data Processing

```python
# loading dataset into a pandas DataFrame
url = 'https://raw.githubusercontent.com/pratimbaidya/Sonar-Rock-vs-Mine-Classification-using-Logistic-Regression/refs/heads/main/sonar.all-data.csv'
sonar_data = pd.read_csv(url , header = 0)
```

```python
sonar_data.head()
```

```python
# number of rows and columns
sonar_data.shape
```

```python
# lets see some statistical measures of the data
sonar_data.describe()
```

```python
# lets check how many rock and mine examples are there in the data
sonar_data['Label'].value_counts()
```

M --> Mine

R --> Rock

```python
# gives the mean for each columns based on the target column (here it is column 60)
sonar_data.groupby('Label').mean()
```

```python
# seperating data and labels
X = sonar_data.drop(columns = 'Label', axis = 1)
y = sonar_data['Label']
```

# Training and Testing data

```python
# Spliting the data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, stratify = y, random_state = 1)
```

```python
# Number of instances in the main data vs training data vs testing data
X.shape[0] , X_train.shape[0], X_test.shape[0]
```

# Modeling --> Logistic Regression

```python
# creating the model
model = LogisticRegression()
```

```python
# training the model with training data
model.fit(X_train, y_train)
```

```python
# lets check the parameters of the fitted model
model.get_params()
```

# Model Evaluation

```python
# accuracy on the training data
train_pred = model.predict(X_train)
train_acu = accuracy_score(train_pred, y_train)
print(f'Accuracy on the training data: {train_acu}')
```

```python
# accuracy on the test data
test_pred = model.predict(X_test)
test_acu = accuracy_score(test_pred, y_test)
print(f'Accuracy on the test data: {test_acu}')
```
